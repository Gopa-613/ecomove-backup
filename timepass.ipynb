```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv('dataset/ts.csv')

# Step 1: Check for missing values
print("Missing Values:\n", df.isnull().sum())

# Step 2: Detect outliers using IQR method and visualize with boxplots
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return outliers

# Visualize boxplots for all numeric columns
plt.figure(figsize=(14, 8))
for i, col in enumerate(df.columns[1:], 1):  # skipping 'year'
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot: {col}")
plt.tight_layout()
plt.show()

# Step 3: Print outliers for each column
print("\nOutliers Detected (IQR method):")
for col in df.columns[1:]:
    outliers = detect_outliers_iqr(df, col)
    if not outliers.empty:
        print(f"{col}:\n", outliers[[col]])
    else:
        print(f"{col}: No outliers")

# Step 4: Optionally remove outliers (can be done per column)
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Create a cleaned copy (example removes outliers from all columns except 'year')
clean_df = df.copy()
for col in clean_df.columns[1:]:
    clean_df = remove_outliers_iqr(clean_df, col)

print("\nDataset shape before outlier removal:", df.shape)
print("Dataset shape after outlier removal:", clean_df.shape)

# Step 5: Feature Scaling (Standardization)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clean_df.drop(columns=['year']))
scaled_df = pd.DataFrame(scaled_features, columns=clean_df.columns[1:])
scaled_df.insert(0, 'year', clean_df['year'].values)

print("\nFinal Scaled Dataset:\n", scaled_df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load dataset
df = pd.read_csv('aqi/petrol.csv')

# Step 1: Check for missing values
print("Missing Values:\n", df.isnull().sum())

# Step 2: Separate numerical and categorical columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Step 3: Visualize outliers using boxplots (for numerical columns)
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot: {col}")
plt.tight_layout()
plt.show()

# Step 4: Outlier Detection with IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return outliers

print("\nOutliers Detected (IQR method):")
for col in numeric_cols:
    outliers = detect_outliers_iqr(df, col)
    if not outliers.empty:
        print(f"{col}:\n", outliers[[col]].head())
    else:
        print(f"{col}: No outliers")

# Step 5: Optionally Remove Outliers (CAUTION: Only do this if you're sure)
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Uncomment below if you want to remove outliers (not recommended for small datasets!)
# for col in numeric_cols:
#     df = remove_outliers_iqr(df, col)

# Step 6: Preprocessing Pipeline
# One-hot encode categorical features, scale numerical ones
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_cols),
    ('cat', OneHotEncoder(drop='first'), categorical_cols)
])

# Fit and transform the entire dataset
processed_data = preprocessor.fit_transform(df)

# Optional: Convert back to DataFrame with column names
# Get new column names for one-hot encoded columns
encoded_cat_cols = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
final_cols = numeric_cols + list(encoded_cat_cols)
processed_df = pd.DataFrame(processed_data, columns=final_cols)

print("\n Final Preprocessed Data:\n", processed_df.head())
```
